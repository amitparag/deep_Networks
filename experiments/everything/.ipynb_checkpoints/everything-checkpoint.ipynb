```python
initial_state = data
```

```python
warmstart_ = warmstart_[:,6:]
```

```python
a = zip(initial_state, warmstart_)
```

```python
model = crocoddyl.ActionModelUnicycle()
for i,j in zip(initial_state, warmstart_):
    u = []
    v = []

    state = j.reshape(30,5)
    i = np.array(i)
    u.append(i)
    xs = state[:,0:3]
    ddp_us = state[:,3:]

    u.extend(_ for _ in xs)
    print(xs.shape)
    
    v.append(np.matrix(_) for _ in ddp_us)
    model.costWeights = np.matrix([1, 0.3]).T
    problem = crocoddyl.ShootingProblem(np.matrix(i).T, [ model ] * 30, model)
    ddp = crocoddyl.SolverDDP(problem)
    #print(i)
    #print(ddp_xs[0])

    ddp.solve(u, v, 1000)
    print(ddp.iter)
```
